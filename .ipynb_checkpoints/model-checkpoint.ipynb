## Config

```python
class Config:   
    # data preprocessing
    validation_fraction = 0.15
    train_batch = 64
    valid_batch = 64
    
    # model setup
    input_ch = 1
    output_ch = 3
    output_features = 2048
    num_classes = 30
    
    # training
    learning_rate = 0.01
    factor = 0.5
    patience = 5
    min_lr = 1e-15
    verbose_lr = True
    epochs = 200
    device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
```

```python
# plot default image
def plot_image(image, ax):
    image = image.reshape(96, 96)
    ax.imshow(image, cmap='gray')
    
    
# plot image with coordinates   
def plot_image_keypoints(image, coords, ax, title):
    plot_image(image, ax)
    ax.scatter(coords[0::2], coords[1::2], marker='x', s=20, color='red')
    plt.title(title)
```

```python
# convert all image values from the pandas dataset into a numpy array, 
# immediately changing the dimension to 96x96x1 and dividing by 255 for normalization
def load_images(df):
    return np.array([np.reshape(np.array(item.split(), dtype=int), (96, 96, 1)) / 255 for item in df['Image']])

# convert all keypoins values from the pandas dataset into a numpy array
def load_keypoints(df):
    return df.drop(columns=['Image']).values

# all train images
train_images = load_images(train)
# rows with missing values (images)
clean_images = load_images(clean_train)
# all train keypoints
train_keypoints = load_keypoints(train)
# rows with missing values (keypoints)
clean_keypoints = load_keypoints(clean_train)
# all test images
test_images = load_images(test)
```

```python
# flip the image and keypoints by a given degree (15 in our case)
def rotate_augmentation(images, keypoints, rotation_angles):
    # lists with flipped images and keypoints
    rotated_images = []
    rotated_keypoints = []
    # we go through the list with corners to flip. 
    # The function will create 2 duplicates of each angle (clockwise and anticlockwise)
    for angle in rotation_angles:
        for angle in [angle, -angle]:
            # we get the rotation matrix M
            M = cv2.getRotationMatrix2D((48,48), angle, 1.)
            # convert the angle into radians and calculate cos, sin
            angle_rad = np.radians(-angle)
            angle_cos, angle_sin = np.cos(angle_rad), np.sin(angle_rad)
            # for each image we make a rotation using the warpAffine function and add it to the list
            rotated_images += [cv2.warpAffine(image, M, (96,96), flags=cv2.INTER_CUBIC) for image in images]
            # for each keypoint, we calculate a new value after rotation and add it to the list
            # based on the two-dimensional rotation matrix
            for keypoint in keypoints:
                rotated_keypoint = keypoint - 48.
                for idx in range(0, len(rotated_keypoint), 2):
                    rotated_keypoint[idx] = rotated_keypoint[idx] * angle_cos - rotated_keypoint[idx+1] * angle_sin
                    rotated_keypoint[idx+1] = rotated_keypoint[idx] * angle_sin + rotated_keypoint[idx+1] * angle_cos
                rotated_keypoint += 48.   
                rotated_keypoints.append(rotated_keypoint)
    # return the finished images, changing the size of the images to 96x96x1        
    return np.reshape(rotated_images, (-1, 96, 96, 1)), rotated_keypoints

rotated_train_images, rotated_train_keypoints = rotate_augmentation(clean_images, clean_keypoints, [15])
# concat the obtained new images and keypoints with the old ones
train_images = np.concatenate((train_images, rotated_train_images))
train_keypoints = np.concatenate((train_keypoints, rotated_train_keypoints))
```

```python
# change the brightness of the image
def change_brightness(images, keypoints):
    # 1 copy increases the brightness of the photo by 25%, the other decreases it by 40%
    # the value must be clipped between [0, 1] due to the fact that normalized data is input
    inc_brightness_images = np.clip(images * 1.25, 0, 1)    
    dec_brightness_images = np.clip(images * 0.6, 0, 1)    
    return np.concatenate((inc_brightness_images, dec_brightness_images)), np.concatenate((keypoints, keypoints))

# concat the obtained new images and keypoints with the old ones
changed_brightness_images, changed_brightness_keypoints = change_brightness(clean_images, clean_keypoints)
train_images = np.concatenate((train_images, changed_brightness_images))
train_keypoints = np.concatenate((train_keypoints, changed_brightness_keypoints))
```

```python
# shift the image to a list of pixels
# each value in the list creates multiple duplicates
def shift_images(images, keypoints, pixel_shifts):
    shifted_images = []
    shifted_keypoints = []
    # we go through the entire shift list and then through all possible shift combinations
    for shift in pixel_shifts:
        for (shift_x, shift_y) in [(-shift, -shift),(-shift, shift),(shift, -shift),(shift, shift)]:
            # shift matrix
            M = np.float32([[1, 0, shift_x], [0, 1, shift_y]])
            for image, keypoint in zip(images, keypoints):
                # shift the image using the warpAffine function and calculate the new keypoints values
                shifted_image = cv2.warpAffine(image, M, (96, 96), flags=cv2.INTER_CUBIC)
                shifted_keypoint = np.array([p + shift_x if i % 2 == 0 else p + shift_y 
                    for i, p in enumerate(keypoint)])
                # check the range of keypoints
                if np.all(0 < shifted_keypoint) and np.all(shifted_keypoint < 96):
                    shifted_images.append(shifted_image.reshape(96,96,1))
                    shifted_keypoints.append(shifted_keypoint)
    # clip the keypoint value and return the result
    shifted_keypoints = np.clip(shifted_keypoints, 0, 96)
    return shifted_images, shifted_keypoints

# concat the obtained new images and keypoints with the old ones
shifted_images, shifted_keypoints = shift_images(clean_images, clean_keypoints, [15])
train_images = np.concatenate((train_images, shifted_images))
train_keypoints = np.concatenate((train_keypoints, shifted_keypoints))
```

## 3 - Dataset and Dataloader

```python
# Splits the dataset into training and validation
X_train, X_val, y_train, y_val = train_test_split(train_images, train_keypoints, test_size=Config.validation_fraction, shuffle=True)
print(f'Train: {X_train.shape[0]}')
print(f'Val: {X_val.shape[0]}')
```

```python
# Create a dataset class with inheritance from torch.utils.data.Dataset
class Facial_Dataset(Dataset):
    def __init__(self, images, keypoints):
        self.images = images
        self.keypoints = keypoints
    
    
    def __len__(self):
        return self.images.shape[0]
    
    
    def __getitem__(self, idx):
        image = self.images[idx]
        keypoint = self.keypoints[idx]
        
        return image, keypoint
```

```python
# create loaders for training and validation sets
train_loader = DataLoader(Facial_Dataset(X_train, y_train), batch_size=Config.train_batch, shuffle=True, num_workers=0)
valid_loader = DataLoader(Facial_Dataset(X_val, y_val), batch_size=Config.valid_batch, shuffle=True, num_workers=0)
```

## 4 - Model (ResNet50 transfer learning)

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        # Conv2d layer to increase the number of channels to 3, 
        # because ResNet accepts 3 channels as input
        self.conv0 = nn.Conv2d(Config.input_ch, Config.output_ch, (1, 1), padding='same')
        # LeakyReLU with negative_slope = 0.1
        self.leaky_relu = nn.LeakyReLU(negative_slope=0.1)
        self.resnet = resnet50()
        # we need everything in resnet50 except the last fully-connected layer, we will replace it with our own
        self.resnet = torch.nn.Sequential(*(list(self.resnet.children())[:-1]))
        # AvgPool2d AvgPool2d layer to get the most important
        self.avg2d = nn.AvgPool2d((1,1))
        # Dropout layer with p = 0.1
        self.dropout = nn.Dropout(0.1)
        # Fully-Connected layer with input 2048 and output 30, since we have 30 different keypoints
        self.fc = nn.Linear(Config.output_features, Config.num_classes)
        
    def forward(self, x):
        # increase the number of channels
        x = self.leaky_relu(self.conv0(x))
        # ResNet50 without FC
        x = self.resnet(x)
        # AvgPool2d + Dropout
        x = self.dropout(self.avg2d(x))
        # flatten along axis=1 to get from (batch_size, 2048, 1, 1) -> (batch_size, 2048)
        x = torch.flatten(x, 1)
        # the last fully-connected layer
        x = self.fc(x)
        return x
```

## 5 - Training the model

```python
def train(model, optimizer, criterion, train_loader, valid_loader, scheduler,
          num_epochs=Config.epochs, valid_loss_min=np.inf, device=Config.device):
    
    # dictionary to store errors during training for plotting
    callbackers = {}
    callbackers['train_loss'] = []
    callbackers['valid_loss'] = []
    
    for e in range(num_epochs):
        train_loss = 0.0
        valid_loss = 0.0
        ###################
        # train the model #
        ###################
        model.train()
        for imgs, keypoints in train_loader:
            # pass the data to the device and reshape images
            imgs, keypoints = imgs.to(device, dtype=torch.float), keypoints.to(device, dtype=torch.float)
            imgs = imgs.view(imgs.shape[0], 1, 96, 96)
            # forward
            out = model.forward(imgs)
            loss = criterion(out, keypoints)
            # optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item() * imgs.shape[0]
        ######################
        # evaluate the model #
        ######################
        model.eval()
        with torch.no_grad():
            targets, predictions = [], []
            for imgs, keypoints in valid_loader:
                # pass the data to the device and reshape images
                imgs, keypoints = imgs.to(device, dtype=torch.float), keypoints.to(device, dtype=torch.float)
                imgs = imgs.view(imgs.shape[0], 1, 96, 96)
                # forward
                out = model.forward(imgs)
                loss = criterion(out, keypoints)
                # we get the result and transfer it to CPU and numpy
                target = keypoints.cpu().numpy()
                prediction = out.cpu().numpy()
                # add the results to the lists
                predictions.extend(prediction)
                targets.extend(target)
                
                valid_loss += loss.item() * imgs.shape[0]
                
        # calculate metrics        
        train_loss /= len(train_loader.sampler)
        valid_loss /= len(valid_loader.sampler)
        # print values
        print(f'Epoch: {e+1} TrainLoss: {train_loss:.3f} ValidLoss: {valid_loss:.3f}')
        # scheduler step
        scheduler.step(valid_loss)
        # save metrics
        callbackers['train_loss'].append(train_loss)
        callbackers['valid_loss'].append(valid_loss)
        
        ######################
        #   save the model   #
        ######################
        if valid_loss <= valid_loss_min:
            torch.save(model.state_dict(), 'model.pt')
            valid_loss_min = valid_loss
        
    return callbackers
```

```python
def visualize_callbackers(callbackers):
    fig, ax =  plt.subplots(1, 1, figsize=(10, 6))
    ax.plot(callbackers['train_loss'], label='Train Loss')
    ax.plot(callbackers['valid_loss'], label='Valid Loss')
    ax.set_title('Losses')
    ax.legend()
    plt.tight_layout()
    plt.show()
```

```python
model = Model().to(Config.device)
optimizer = optim.Adam(model.parameters(), lr=Config.learning_rate)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=Config.factor, patience=Config.patience, 
                                                 min_lr=Config.min_lr, verbose=Config.verbose_lr)
criterion = nn.MSELoss()
callbackers = train(model, optimizer, criterion, train_loader, valid_loader, scheduler)
visualize_callbackers(callbackers)
```

## 6 - Get test predictions and make submission

```python
def predict(model, criterion, images, device=Config.device):
    model.eval()
    with torch.no_grad():
        # convert to FloatTensor and resize to (m, 1, 96, 96)
        images = torch.FloatTensor(images)
        images = images.view(images.shape[0], 1, 96, 96)
        # forward
        out = model.forward(images)
    return out.cpu().numpy()
```

```python
# create a new model object (which will be on the CPU for the prediction)
model = Model()
# load the latest best model
model.load_state_dict(torch.load('model.pt'))
# get test predictions
test_predictions = predict(model, nn.MSELoss(), test_images)
```

```python
feature_names = list(ids['FeatureName'])
image_ids = list(ids['ImageId'] - 1)
row_ids = list(ids['RowId'])

# feature name indexes
feature_list = [feature_names.index(feature) for feature in feature_names]
predictions = [test_predictions[x][y] for x, y in zip(image_ids, feature_list)]


row_ids = pd.Series(row_ids, name='RowId')
locations = pd.Series(predictions, name='Location')
# make predictions in the segment [0, 96]
locations = locations.clip(0, 96)
result = pd.concat([row_ids, locations], axis = 1)
result.head(5)
```
